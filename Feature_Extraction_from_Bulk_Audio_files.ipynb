```python
import os
import librosa
import numpy as np
import csv
import pandas as pd


csvname = "/content/drive/MyDrive/Datasets/FE2-file.csv"

coloumn_1 = ['file-name', 'rms']

chroma_stft_col = []
for i in range(0, 12):
  chroma_stft_col.append('chroma_stft{i}'.format(i=i+1))

coloumn_2 = ['spec_centroid', 'spec_bandwidth', 'rolloff', 'zrc']

mfcc_col = []
for i in range(0,20):
  mfcc_col.append('mfcc{i}'.format(i=i+1))

chroma_cgt_col = []
for i in range(0, 12):
  chroma_cgt_col.append('chroma_cgt_col{i}'.format(i=i+1))

chroma_cens_col = []
for i in range(0, 12):
  chroma_cens_col.append('chroma_cens_col{i}'.format(i=i+1))

melspec_col = []
for i in range(0, 128):
  melspec_col.append('melspec_col{i}'.format(i=i+1))

contrast_col = []
for i in range(0, 7):
  contrast_col.append('contrast_col{i}'.format(i=i+1))

coloumn_3 = ['flatness']

tonnetz_col = []
for i in range(0, 6):
  tonnetz_col.append('tonnetz_col{i}'.format(i=i+1))

coloumns_final = coloumn_1 + chroma_stft_col + coloumn_2 + mfcc_col + chroma_cgt_col + chroma_cens_col + melspec_col + contrast_col + coloumn_3 + tonnetz_col + ['LABEL']


with open(csvname, 'a') as csvfile:
  csvwriter = csv.writer(csvfile)
  csvwriter.writerow(coloumns_final)


filename = '/content/drive/MyDrive/Complete Data Coswara/DataSet'
labels = ['healthy', 'positive_asymp', 'positive_mild', 'positive_moderate']
# Changing the directory to where the dataset is available 
os.chdir(filename)
for i in labels:
  path = filename+'/'+i   
  folders_list = os.listdir(path)
  for j in range(len(folders_list)):
    path2 = path+'/'+folders_list[j]
    wavfiles_list = os.listdir(path2)
    for k in range(len(wavfiles_list)):
      path3 = path2+'/'+wavfiles_list[k]
      
      if not path3.endswith('.wav'):
        continue
      x, sr = librosa.load(path3)


      rmse = np.mean(librosa.feature.rms(y=x))   

      chroma_stft = librosa.feature.chroma_stft(y=x, sr=sr)

      spec_cent = np.mean(librosa.feature.spectral_centroid(y=x, sr=sr)) 

      spec_bw = np.mean(librosa.feature.spectral_bandwidth(y=x, sr=sr))  

      rolloff = np.mean(librosa.feature.spectral_rolloff(y=x, sr=sr)) 

      zcr = np.mean(librosa.feature.zero_crossing_rate(x)) 

      mfcc = librosa.feature.mfcc(y=x, sr=sr)

      chroma_qt = librosa.feature.chroma_cqt(y=x, sr=sr)

      chroma_cens = librosa.feature.chroma_cens(y=x, sr=sr)

      melspec = librosa.feature.melspectrogram(y=x, sr=sr)

      contrast = librosa.feature.spectral_contrast(y=x, sr=sr)

      flatness = np.mean(librosa.feature.spectral_flatness(y=x))

      tonnetz = librosa.feature.tonnetz(y=x, sr=sr)

      file_name = folders_list[j] + '_' + wavfiles_list[k] 

      rows1 = [file_name, rmse]
      # chroma stft col 
      chroma_stft_lst = [] 
      for l in range(0, 12):
        chroma_stft_lst.append(np.mean(chroma_stft[l]))
      row2 =  [spec_cent, spec_bw, rolloff, zcr]  
      # mfcc 
      mfcc_lst = [] 
      for l in range(0,20):
        mfcc_lst.append(np.mean(mfcc[l]))
      # chroma cq 
      chroma_cq_lst = [] 
      for l in range(0, 12):
        chroma_cq_lst.append(np.mean(chroma_qt[l]))
      # chroma cens 
      chroma_cens_lst = [] 
      for l in range(0, 12):
        chroma_cens_lst.append(np.mean(chroma_cens[l]))
      # mel spectral
      mel_spectral_lst = []
      for l in range(0, 128):
        mel_spectral_lst.append(np.mean(melspec[l]))
      # contrast 
      contrast_lst = []
      for l in range(0, 7):
        contrast_lst.append(np.mean(contrast[l]))
      rows3 = [flatness]
      # tonnentz
      tonnentz_lst = []
      for l in range(0, 6):
        tonnentz_lst.append(np.mean(tonnetz[l]))
      
      
      label = i

      rows_final = rows1 + chroma_stft_lst + row2 + mfcc_lst + chroma_cq_lst + chroma_cens_lst + mel_spectral_lst + contrast_lst + rows3 + tonnentz_lst + [label]


      with open(csvname, 'a') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow(rows_final)
```
