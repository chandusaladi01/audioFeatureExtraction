```python
import numpy as np
import librosa

path = '/content/1_Negativo_34_f_17-6-20_1.wav'

x, sr = librosa.load(path)


rmse = (librosa.feature.rms(y=x))   

chroma_stft = librosa.feature.chroma_stft(y=x, sr=sr)

spec_cent = (librosa.feature.spectral_centroid(y=x, sr=sr)) 

spec_bw = (librosa.feature.spectral_bandwidth(y=x, sr=sr))  

rolloff = (librosa.feature.spectral_rolloff(y=x, sr=sr)) 

zcr = (librosa.feature.zero_crossing_rate(x)) 

mfcc = librosa.feature.mfcc(y=x, sr=sr)

chroma_qt = librosa.feature.chroma_cqt(y=x, sr=sr)

chroma_cens = librosa.feature.chroma_cens(y=x, sr=sr)

melspec = librosa.feature.melspectrogram(y=x, sr=sr)

contrast = librosa.feature.spectral_contrast(y=x, sr=sr)

flatness = (librosa.feature.spectral_flatness(y=x))

tonnetz = librosa.feature.tonnetz(y=x, sr=sr)


print('rmse :', rmse, rmse.shape)
print('chroma_stft :', chroma_stft, chroma_stft.shape)
print('spec_cent :', spec_cent, spec_cent.shape)
print('spec_bw :', spec_bw, spec_bw.shape)
print('rolloff :', rolloff, rolloff.shape)
print('zcr :', zcr, zcr.shape)
print('mfcc :', mfcc, mfcc.shape)
print('chroma_qt :', chroma_qt, chroma_qt.shape)
print('chroma_cens :', chroma_cens, chroma_cens.shape)
print('melspec :', melspec, melspec.shape)
print('contrast :', contrast, contrast.shape)
print('flatness :', flatness, flatness.shape)
print('tonnetz :', tonnetz, tonnetz.shape)
```
